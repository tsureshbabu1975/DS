<h3> Q5: Find the closest points</h3>

consider you have given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P
<br>cosine distance between two points (x,y) and (p,q) is defind as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)
<img src='https://i.imgur.com/vIFPOcG.jpg', width=300>
Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
import math

# function to calculate cosine distance between two points
# points from plane x=(x,y) and query point P=(p,q)
# x is tuple x=(x,y)=(x[0],x[1])  
# P is also tuple p=(p,q)=(p[0],[1])

# cosine distance formula = cos-1[(x.p+y.q)/ (SQRT(x**2+y**2).SQRT(p**2+q**2))]
def cosine_distnace(x,p):
    return math.acos(((x[0]*p[0])+(x[1]*p[1]))/
                          ((math.sqrt((x[0]**2)+(x[1]**2)))*(math.sqrt((p[0]**2)+(p[1]**2)))))

#find the n points closest to query point
def closest_points_to_p(S, P, n):
    #Invoke cosine distance function by passing each point and query point
    cosine_dist_lst = [cosine_distnace(i,P) for i in S]
    #merge plane points with cosine distance and create disctionary
    #sort the dictionary based on cosine distanct
    closest_points = sorted(dict(zip(cosine_dist_lst,S)).items())
    #return first n points 
    return closest_points[:n]
    #return closest_points_to_p  # its list of tuples

#Sample plane point
S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]

#Sample query point
P= (3,-4)

# n is number of closest points to query point to be identified
n=5

#call closest_points_to_p function and unzip the distance and points
closest_points_to_p = list(zip(*closest_points_to_p(S, P,n)))

#print points closer to query point
print('These points {} are closer to the query point {}'.format(closest_points_to_p[1:],P))
```

```python
import re
Red= [(1,1),(2,1),(4,2),(2,4), (-1,4)]
Blue= [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]
Lines=["1x+1y+0","1x-1y+0","1x+0y-3","0x+1y-0.5"]

def mul_dp_lci(p,lci):
    #Example: line: 1x+1y+0 and point(1,1)
    #1*1+1*1+0 = 2
    return ((lci[0]*p[0])+(lci[1]*p[1])+lci[2])

#function to find group of datapoints1 and datapoints2 are seperated or not 
#dp1= datapoints 1[Example: Red]
#dp2 = datapoints 2 [Ex: Blue]
#lci = line coefficient and intercepts 
def are_datapoints_separated(dp1,dp2,lci):
    #invoke data points multiplication with line
    #if the value is +ve then set True else False
    dp1_T_F = list(map(lambda x:(True if x>0 else False),
                                 list([mul_dp_lci(p,lci) for p in dp1])))
    dp2_T_F = list(map(lambda x:(True if x>0 else False),
                                 list([mul_dp_lci(p,lci) for p in dp2])))
    
    #check all datapoints are in one side or not. if so, SET as True 
    #ie all datapoints should be either "True" or "False" but not mixed.
    #all function returns true only if all points are same(True / False)
    dp1_all_in_one =all(element == dp1_T_F[0] for element in dp1_T_F)
    dp2_all_in_one =all(element == dp2_T_F[0] for element in dp2_T_F)
    
    #if both data points True then it is seperated clearly else No
    if (dp1_all_in_one==True and dp2_all_in_one==True):
        return 'Yes'
    else:
        return 'No'

    return 
    
#function to extract coefficients and intercepter from the plane /line
def extract_coeff_intercepter_from_line(lines):
    #Extract line coefficients and intercept using re function
    #It is important to extract coefficient and interceptwith sign to 
    #understand the points are in whcih side
    coeff_str=[re.findall(r'[\d\.\-\+]+', i) for i in Lines]
    #Convert them into float numbers
    coeff =[list(map(lambda x:float(x),x)) for x in coeff_str]
    return coeff

#driver program
#call extract function to seperate coeff and intercepts
coeff=extract_coeff_intercepter_from_line(Lines)

#invoke find datapoint side of the plane of given 
line_separates_yes_no = [are_datapoints_separated(Red,Blue,linecoeffintercept) for linecoeffintercept in coeff]

#zip lines and line_separates_yes_no 
l_seperates = zip(Lines,line_separates_yes_no)
for l in l_seperates:
    print('Does Line {} seperate Red and Blue points ?--> {}'.format(l[0],l[1]))
```
