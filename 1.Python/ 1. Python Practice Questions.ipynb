```python
'''Exercise 1: 
   Write a function that inputs a number and prints the multiplication table of that number'''

#table size is fixed 
table_size =10
#get the input from the user
table_number=input('Please type the number to generate table \n')

#function to validate the number is numeric or not. 
def is_valid_number(num):
    if num.isnumeric()==False:
        print("This is not a number!!!. Please enter only numbers")
        return False
    return True

#Generate table output for the given table number. 
if is_valid_number(table_number):
    #using range function generate list from 1 to table size.
    lst = range(1,table_size+1)
    
    #using map and lambda function generate table output and store it in new_lst
    new_lst = list(map(lambda x: x * int(table_number), lst))
    
    #printing
    print('\n\nMultiplication table for the given number {} '.format(table_number))
    
    #enumerate starts from zero, hence add 1 start the multiplication table from 1
    for idx,val in enumerate(new_lst):
        print('{} X {} = {} '.format(idx+1,table_number,val))
```

```python
'''
Exercise 2: 
Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes

Expected o/p is:
(3, 5), (5, 7), (11, 13), (17, 19), (29, 31), (41, 43), (59, 61), (71, 73),... less than 1000

Approach: There are multiple approaches to find prime numbers. 

Method 1: Brute-force method
    In this method, test all divisors from 2 to n-1 and skip 1 and n. 
    1.If the integer is less than equal to 1, it returns False.
    2.If the given number is divisible by any of the numbers from 2 to n, the function will return False
    3. Else it will return True
Time complexity is high for large/very large numbers. Hence this approach is dropped. 


Method 2: Mirror Factorization (Divisible reduction) Technique
    The square root of the number divides the factors into two halves.
    Example: 36 =>   =1*36  | =2*18    | =3*12  | =4*9
                        -----------  =6*6 ------------
                     =9*4   | =12*3    | =18*2  | =36*1

    Hence a simple hack by reducing the number of divisors to SQRT of the number.  
    if the number is a perfect square, we can shift the line by 1. 
    (Use Math module to find the floor and add 1)   
    and if we can get the integer value of the line which divides
    
    In this method, the time complexity has reduced drastically.
    However we are still iterating through all numbers. 
    Except 2, all even numbers are not prime numbers, Hence, will remove even numbers
    and iterate through with odd numbers.this would reduce the time complexity further.

Method 3: Sieve Method
    This method prints all the primes smaller than or equal to a given number n.
    This method is considered to be the most efficient method to generate 
    all the primes /list of numbers smaller than the given number n. 
    Please note that, this method is not suitable to check for a "particular prime number"
    Time Complexity of the code would take O(N log (log N))
    
Method 4: Modified Sieve Method - Time Complexity o(n)***********
    Algorithm works as below:
    For every number i where i varies from 2 to n-1: 
    Check if the number is prime. If the number is prime, store it in prime array.

    For every prime numbers j less than or equal to the smallest prime factor p of i:
    Mark all numbers j*p as non_prime.
    Mark smallest prime factor of j*p as j    
    
Since, our problem is to generate list of twin primes lesser than 1000, 
time complexity is not significant, However, higher order number generation would impact. 
Hence, most suited method is Modified Sieve Method

Metrics: for 100k prime generation: Method 4 takes about 0.11138129234313965 seconds.
Where as same executed with Method 3 it took around 0.018650054931640625 seconds.
executed with 16GB RAM, i7 8th Gen processor. 
In addition to this, there are many other applications are running in parallel.
This could vary with laptop configuration and number of applications installed.


Analysis of Different Methods to find Prime Number in Python:
https://www.geeksforgeeks.org/analysis-different-methods-find-prime-number-python/

code reference: https://www.geeksforgeeks.org/sieve-eratosthenes-0n-time-complexity/

However is modified to work for twin primes
'''
# Modified Sieve Method code
import time
# The below program designed to work / generate 
# prime number lesser than MAX_SIZE specified. 
# Change the MAX_SIZE as per your need.
MAX_SIZE = 1000

# isPrime[] : isPrime[i] is true if number is prime 
# prime[] : stores all prime number less than n 
# SPF[] that store smallest prime

# For ex : smallest prime factor of '8' and '16' is '2' 
# Hence, SPF[8] = 2 ,and SPF[16] = 2 ] 

#Initialize all isprime is true with size of MAX_SIZE.
isprime = [True] * MAX_SIZE 

#similarly initialize SPF is NONE for all
SPF = [None] * (MAX_SIZE) 

#Empty prime list
prime = [] 

# function to generate all prime number 
def manipulated_seive(n): 
    # 0 and 1 are not prime. Hence set as "False"
    isprime[0] = isprime[1] = False
    
    # Iterate less than n and start with 2 
    for i in range(2, n): 
        # If isPrime[i] == True then i is prime number 
        if isprime[i] == True: 
            # Append prime[] vector 
            prime.append(i) 
            # A prime number is its own smallest prime factor 
            SPF[i] = i 

        # Remove all multiples of i*prime[j]. These are NOT primes 
        # Prime[i * prime[j]] = false and put 
        # smallest prime factor of i*Prime[j] 
        # as prime[j] [ for exp :let i = 5 , j = 0 , 
        # prime[j] = 2 [ i*prime[j] = 10 ] 
        # so smallest prime factor of '10' is '2' 
        # that is prime[j] ] this loop run only one 
        # time for number which are not prime 
        j = 0
        
        while (j < len(prime) and 
               i * prime[j] < n and
               prime[j] <= SPF[i]): 
            
            isprime[i * prime[j]] = False
            # put smallest prime factor of i*prime[j] 
            SPF[i * prime[j]] = prime[j] 
            #increment J by 1
            j += 1
  
    #The above code identifies all the prime code. 
    
    #Construct twin prime number using prime[i] and isprime[]=True
    #for Odd Twin numbers, 2 is not considered. Hence, SET isprime[2]=False
    isprime[2]=False 
    
    # Twin prime numbers. The below loop can be optimized further?
    twinprime_lst=[]
    for p in range(2, n-1): 
        if isprime[p] and isprime[p + 2]:
            twinprime_lst.append((tuple([p,p+2])))
        
    return twinprime_lst   


# Driver function 
t0 = time.time() 
twin_primes = manipulated_seive(MAX_SIZE) 
print("Total prime numbers in range:", twin_primes) 
  
t1 = time.time() 
print("Time required:", t1 - t0) 
```

```python
'''
Exercise: 3
Write a program to find out the prime factors of a number. Example: prime factors of 56 -
2, 2, 2, 7

Approach: we have to iterate two loops, first loop for 2 prime factor. 
second loop from 3 sqrt(remaining number) with step size of 2 minimise the number of iterations.

'''
# Math module has floor function
import math
#import time for computational time
import time
# Python Program to find prime factors of a given number
def primeFactors(n): 
    # empty list
    primeFactors_lst=[]  
    # Step 1: divide the given number with lowest prime factor of 2 till you get reminder >0
    # loop through till we it is not divisible by 2
    while n % 2 == 0: 
        #since the given number is divisible by 2, add it to prime factor list
        primeFactors_lst.append(2), 
        # update the quotient in n itself 
        n = n / 2
    #if n=1 then no need to loop further, return 
    #example 8 = 2*2*2 no need to check further
    if n==1:
        return primeFactors_lst
    #Step 2: if n is >2 and add loop through for add numbers 
    #till sqrt.of remaining quotient which is updated in n itself
    else:
        # start the loop with 3 and step size as 2
        # also apply floor of squareroot of n. And add 1
        for i in range(3,math.floor(math.sqrt(n))+1,2): 
            # while i divides n , add it to prime number list 
            while n % i== 0: 
                primeFactors_lst.append(i) 
                #update the quotient to n itself to iterate further
                n = n / i 
              
    # final n must be prime and add it to prime_list
    primeFactors_lst.append(int(n)) 
    return primeFactors_lst

#any integer as input
n = 10000056000020 
# invoke prime factor function set the timer on
t0 = time.time() 
primeFactors_lst= primeFactors(n) 
#time to just execute only. printing is not considered for execution
t1 = time.time() 
#print prime factors
print("prime factors of a given number {} are ({}):".format(n,primeFactors_lst)) 
#print the time taken to execute

print("Time required:", t1 - t0) 
```

```python
'''Exercise 4: 
Write a program to implement these formulae of permutations and combinations.
Number of permutations of n objects taken r at a time: p(n, r) = n! / (n-r)!. Number of
combinations of n objects taken r at a time is: c(n, r) = n! / (r!*(n-r)!) = p(n,r) / r!


Approach: 
    permutations: p(n,r)=n!/(n-r)!: 
    combinations: c(n,r) = n!/(r!*(n-r)! : 
'''

#function to validate the number is numeric or not. 
def is_valid_number(num):
    if num.isnumeric()==False:
        print("This is not a number!!!. Please enter only numbers")
        return False
    return True

def factorial(n):
    # use recursive funtion to calculate factorial
    return 1 if (n==1) else n * factorial(n - 1);  

def npr(n,r):
    #npr = n!/(n-r)!. To implement, call factorial function
    return int(factorial(n)/factorial(n-r))

def ncr(n,r):
    #ncr = n!/(r!*(n-r)!. or npr/r! 
    #To implement, call npr and factorial function
    return int(npr(n,r)/factorial(r))

#get the input from the user
n=input('Please enter total number of objects in the set (n) \n')
r=input('Please specify number of choosing objects from the set (r) \n')

#if both n and r valid numbers then proceed, else throw exception and stop
if is_valid_number(n) and is_valid_number(r):
    n = int(n)
    r = int(r)
    if (n==0 or r==0):
        print('Either n or r is zero.To compute permutation and combination n and r to be greater than zero!')
    else:
        #compute permutation
        p = npr(n,r)
        #compute combination
        c = ncr(n,r)
        #print npr and ncr 
        print(" \n Permutations, p({}, {}) = {}".format(n,r,p)) 
        print(" Combinations c({}, {}) = {}".format(n,r,c))
```

```python
'''Exercise 6:
Write a function cubesum() that accepts an integer and returns the sum of the cubes of
individual digits of that number. Use this function to make functions PrintArmstrong() and
isArmstrong() to print Armstrong numbers and to find whether is an Armstrong number.
'''
from functools import reduce

#function to validate the number is numeric or not. 
def is_valid_number(num):
    if num.isnumeric()==False:
        print("This is not a number!!!. Please enter only numbers")
        return False
    return True

def CubeDigitSum(n):
    # convert number to string iterables
    s =str(n)
    #calculate cube of each digit 
    cube_lst=[int(i)**3 for i in s]
    #calculate sum
    return  reduce(lambda a,b: a+b,cube_lst)

def isArmstrong(n):
    return True if n==CubeDigitSum(n) else False

def PrintArmstrong():
    # 3 digit number Armstrong numbers 
    # loop to check between 100 to 999
    armstrongs_lst=[]
    for n in range(100,1000):
        a = CubeDigitSum(n)
        if n==a:
            armstrongs_lst.append(a)
    print('\n3 digit armstrong numbers are {}'.format(armstrongs_lst))

#main

#get the input from the user
n=input('Please enter a number to verify is Armstrong Number or not \n')
# validate the input number 
if is_valid_number(n) and int(n)>0:
    n=int(n)
    s = CubeDigitSum(n)

    if isArmstrong(n):
        print('\nThe given number {} and Cube digits sum {} are equal. \
Hence it is an Armstrong Number.'.format(n,s))
    else:
        print('\nThe given number {} and Cube digits sum {} are not equal. \
Hence it is not an Armstrong Number.'.format(n,s))
    PrintArmstrong()
```

```python
'''
Exercise 8:
If all digits of a number n are multiplied by each other repeating with the product, the one
digit number obtained at last is called the multiplicative digital root of n. The number of
times digits need to be multiplied to reach one digit is called the multiplicative
persistance of n.
Example: 86 -> 48 -> 32 -> 6 (MDR 6, MPersistence 3)
 341 -> 12->2 (MDR 2, MPersistence 2)
Using the function prodDigits() of previous exercise write functions MDR() and
MPersistence() that input a number and return its multiplicative digital root and
multiplicative persistence respectively
'''

from functools import reduce

#function to validate the number is numeric or not. 
def is_valid_number(num):
    if num.isnumeric()==False:
        print("This is not a number!!!. Please enter only numbers")
        return False
    return True

#function to calculate product of digits
def prodDigits(n):
    # convert number to string iterables
    s =str(n)
    #product of digit using reduce function
    return reduce(lambda a,b: int(a)*int(b),s)

def MDR(n):
    if n<10:
        return n
    else:
        #recursive function
        return MDR(prodDigits(n))

def MPersistence(n):
    counter=0
    if len(str(n)) > 1 :
        counter += 1
        while True :
            p=prodDigits(n)
            if len(str(p)) > 1:
                counter += 1
                n=p
            else:
                return counter  #break the loop and exit the function
      
        
#get the input from the user
n=input('Please enter a number to get product of digits! \n')
# validate the input number 
if is_valid_number(n) and int(n)>0:
    n=int(n)
    #invoke MDR function
    m=MDR(n)
    #print MDR 
    print('{} is Multiplicative digital root (MDR)  of a given number {}'.format(m,n))
    
    #invoke MPersistence function
    c=MPersistence(n)
    #print MPersistence 
    print('{} MPersistence is a given number {}'.format(c,n))
        
```

```python
'''Exercise 9:
Write a function sumPdivisors() that finds the sum of proper divisors of a number. Proper
divisors of a number are those numbers by which the number is divisible, except the
number itself. For example proper divisors of 36 are 1, 2, 3, 4, 6, 9, 18

'''
import math  
from functools import reduce

#function to validate the number is numeric or not. 
def is_valid_number(num):
    if num.isnumeric()==False:
        print("This is not a number!!!. Please enter only numbers")
        return False
    return True

#function to generate divisors 
def generateDivisors(n) : 
    divisor_lst=[]  
    # Note that this loop runs from 2 to till square root 
    for i in range(2,math.floor(math.sqrt(n))+1): 
        if (n % i == 0) : 
            # If divisors are equal, print only one 
            if (n / i == i) : 
                divisor_lst.append(i)
            else : 
                # Otherwise print both 
                divisor_lst.append(i)
                divisor_lst.append(math.floor(n/i))
    #one also is a divisor 
    divisor_lst.append(1)
    return divisor_lst

def sumPDivisors(lst):
    #use reduce 
    return reduce((lambda a,b: a+b),lst)

#get the input from the user
n=input('Please enter a number to get all the divisors! \n')
# validate the input number 
if is_valid_number(n) and int(n)>0:
    n=int(n)
    #call generateDivisors to get list of divisors. 
    #list is unsorted. hence sort it before printing
    divisor_lst=generateDivisors(n) 
    if len(divisor_lst)==1:
        print ("The given number {} is prime!. Hence 1 and same number {}\
 are the divisors.".format(n,n))
    else:
        print ("The divisors of {} are: {}".format(n,sorted(divisor_lst)))
        # pass the divisior list to sumPDivisors to get sum of divisors
        s= sumPDivisors(divisor_lst)
        print ("The proper divisors sum is: {}".format(s))
```

```python
'''Exercise 11:
Two different numbers are called amicable numbers if the sum of the proper divisors of
each is equal to the other number. For example 220 and 284 are amicable numbers.

Sum of proper divisors of 220 = 1+2+4+5+10+11+20+22+44+55+110 = 284
Sum of proper divisors of 284 = 1+2+4+71+142 = 220
Write a function to print pairs of amicable numbers in a range
'''

import math  
from functools import reduce

#function to validate the number is numeric or not. 
def is_valid_number(num):
    if num.isnumeric()==False:
        print("This is not a number!!!. Please enter only numbers")
        return False
    return True

#function to generate divisors 
def generateDivisors(n) : 
    divisor_lst=[]  
    # Note that this loop runs from 2 to till square root 
    for i in range(2,math.floor(math.sqrt(n))+1): 
        if (n % i == 0) : 
            # If divisors are equal, print only one 
            if (n / i == i) : 
                divisor_lst.append(i)
            else : 
                # Otherwise print both 
                divisor_lst.append(i)
                divisor_lst.append(math.floor(n/i))
    #one also is a divisor 
    divisor_lst.append(1)
    return divisor_lst

def sumPDivisors(lst):
    #use reduce 
    return reduce((lambda a,b: a+b),lst)

#get the input from the user
f=input('Please enter first number to check is amicable number ! \n')
s=input('Please enter second number ! \n')
# validate the input number 
if int(f)== int(s):
    print("The given numbers {},{} are same!. Please try with different numbers".format(f,s))
    
elif int(f)>0 and is_valid_number(f) and int(s)>0 and  is_valid_number(s):
    f=int(f)
    s=int(s)
    #call generateDivisors to get list of divisors. 
    #list is unsorted. hence sort it before printing
    f_divisor_lst=generateDivisors(f) 
    s_divisor_lst=generateDivisors(s)
    
    #check if first number is prime 
    if len(f_divisor_lst)==1:
        print ("The given first number {} is prime! and can't be amicable number \
with {}".format(f,s))
    #check if second number is prime 
    elif len(s_divisor_lst)==1:
        print ("The given second number {} is prime! and can't be amicable number \
with {}".format(s,f))
    #otherwise both first and second numbers are different. 
    #print first and second numbers divisables 
    else:
        # pass the divisior list to sumPDivisors to get sum of divisors
        f_divisor_sum= sumPDivisors(f_divisor_lst)
        s_divisor_sum= sumPDivisors(s_divisor_lst)

        print ("The first number {} and it's sum of divisors: \
{}".format(f,f_divisor_sum))
        print ("The second number {} and it's sum of divisors: \
{}".format(s,s_divisor_sum))
       
        # first number equals with sum of second number divisors  
        # second number equals with sum of first number divisors
        if (f==s_divisor_sum) and (s==f_divisor_sum) :
            print ("It is amicable number!. first number {} \
and sum of second number divisors {} matches. Likewise, \
second number {} and sum of first number divisors {} \
matches".format(f,s_divisor_sum,s,f_divisor_sum))
        else:
            print ("It isn't amicable number!. Either first number {} \
and sum of second number divisors {} doesn't match. Or \
second number {} and sum of first number divisors {} \
doesn't match".format(f,s_divisor_sum,s,f_divisor_sum))
```
